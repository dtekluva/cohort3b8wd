```python
target_country = "Nigeria"
filtered_target_country = melted_data_cases[melted_data_cases["Country/Region"] == target_country]

rolling_diff_cases = filtered_target_country["cases"].diff()
rolling_diff_deaths= filtered_target_country["deaths"].diff()
filtered_target_country
```

```python
filtered_target_country["rolling_cases"] = rolling_diff_cases
filtered_target_country["rolling_deaths"] = rolling_diff_deaths
filtered_target_country
```

```python
average_cases = rolling_diff_cases.max()
average_cases
```

```python
filtered_target_country["dates"] = filtered_target_country.dates.astype("datetime64")
```

```python
filtered_target_country.info() 
```

```python
filtered_target_country["day"] = filtered_target_country.dates.dt.day_name()
```

```python
filtered_target_country
```

```python
filtered_target_country.groupby("day").mean().sort_values("rolling_cases", ascending=False
                                                         )
```

```python
day_groups = filtered_target_country.groupby("day").mean().sort_values("rolling_cases", ascending=False
                                                         ).reset_index()
```

```python
day_groups.plot(x = "day", y = "rolling_cases", kind = "bar")
```
