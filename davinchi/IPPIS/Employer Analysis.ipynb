```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel("IPPIS_DEDUCTION_REPORT1.xlsx", sheet_name = None)
data = pd.concat(data[frame] for frame in data.keys())

analysis = data.groupby(['Ministry Name', 'Reporting Month']).agg(["sum","count"]).reset_index()
```

```python
analysis
```

```python
analysis_columns = analysis.columns.map(''.join)
analysis.columns = analysis_columns

analysis
```

```python
required_columns = ['Ministry Name',
                    'Reporting Month',
                    'Employee Numbercount',
                    'Deduction Amountsum',
                    'Deduction Amountcount',
                    
                    ]

target = analysis[required_columns]
analysis = target.rename(columns = {'Employee Numbercount': 'Employee Count', 'Deduction Amountsum': 'TotalDeducted', 'Deduction Amountcount': 'Nos Deducted'})
analysis
```

```python
analysis2 = analysis.groupby(['Ministry Name']).agg(["sum","count"]).reset_index()
```

```python
analysis_columns = analysis2.columns.map(''.join)
analysis2.columns = analysis_columns
analysis2
```

```python
# analysis2.to_excel("Ministry_Analysis.xlsx", index=False)
```

```python
analysis2.columns
```

```python
required_columns = ['Ministry Name', 
                    'Employee Countsum', 
                    'Employee Countcount',
                    'TotalDeductedsum'                
                   ]
target = analysis2[required_columns]
analysis2 = target.rename(columns = {'Employee Countsum': 'Employee Count', 'TotalDeductedsum': 'TotalDeducted', 'Employee Countcount': 'MonthsDeductedCount'})
analysis2
```

```python
finished = analysis2.sort_values('MonthsDeductedCount', ascending=False)
finished.to_excel('Ministry_Analysis.xlsx', index=False)
```
